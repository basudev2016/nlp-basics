```python
import pandas as pd
```

```python
df = pd.read_csv('D:/Nitte_NLP/nlp-basics/Day-1/2_Text-preprocessong/cleaned_text.csv')
df.head()
```

```python
df['lemmatized_text'] = df['lemmatized_tokens'].apply(lambda x: ' '.join(x))
df
```

```python
# BOW
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

# Fit and transform
X_bow = vectorizer.fit_transform(df['lemmatized_text'])

print("\nShape of BoW Matrix:", X_bow.shape)
print("\nFirst Row as Array:", X_bow[0].toarray())
print("\nVocabulary Sample:", vectorizer.get_feature_names_out()[:10])
```

```python
#TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()

X_tfidf = tfidf_vectorizer.fit_transform(df['lemmatized_text'])

print("\nShape of TF-IDF Matrix:", X_tfidf.shape)
print("\nFirst Row as Array:", X_tfidf[0].toarray())
print("\nVocabulary Sample:", tfidf_vectorizer.get_feature_names_out()[:10])
```

```python
# TF IDF HeatMap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# First, create a DataFrame from the sparse TF-IDF matrix
# (use only the first 10 words to keep the heatmap small and readable)
tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# For a clean small heatmap, take only top 10 words from vocabulary
selected_words = tfidf_vectorizer.get_feature_names_out()[:10]

# Slice only those columns
small_tfidf_df = tfidf_df[selected_words]

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(small_tfidf_df, annot=True, cmap="YlGnBu", cbar=True)
plt.title("TF-IDF Heatmap (Top 10 Words)")
plt.xlabel("Words")
plt.ylabel("Documents")
plt.show()
```

```python
# Create DTM from Bag of Words
dtm = pd.DataFrame(X_bow.toarray(), columns=vectorizer.get_feature_names_out())

print("\nDocument-Term Matrix (DTM):")
print(dtm.head())
```

```python
# Create TDM by transposing the DTM
tdm = dtm.transpose()

print("\nTerm-Document Matrix (TDM): ",tdm)
```

```python
# Sparse Matrix Visualization
import matplotlib.pyplot as plt

# Visualize sparsity for BoW matrix
plt.figure(figsize=(10, 6))
plt.spy(X_bow, markersize=5)
plt.title('Sparsity Pattern of Bag of Words Matrix')
plt.xlabel('Words')
plt.ylabel('Documents')
plt.show()

# (Optional) Visualize sparsity for TF-IDF matrix
plt.figure(figsize=(10, 6))
plt.spy(X_tfidf, markersize=5)
plt.title('Sparsity Pattern of TF-IDF Matrix')
plt.xlabel('Words')
plt.ylabel('Documents')
plt.show()
```
